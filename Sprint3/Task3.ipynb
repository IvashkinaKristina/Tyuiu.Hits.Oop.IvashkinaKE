```
public interface IMaxWithdrawalamount
{
    int MaxWithdrawalamount {get; set;}
    void GetInfo();
}

public interface IWithdrawMoney
{
    int WithdrawMoney {get; set;}
}


public class BankAccount
{
   public int AccountNumber {get; private set;}
   public decimal Balance {get; set;}
   public string AccountType {get; set;}
   public string UserName { get; set; }
   public bool Active { get; set; }
   public DateTime OpenDate { get; private set; } // Дата открытия счета
   /*public List<Transaction> Transactions { get; private set; } // Список транзакций по счету*/
   
   public BankAccount (int numb, decimal balance, string accountType, string userName, bool active, DateTime openDate)
   {
      AccountNumber = numb;
      Balance = balance;
      AccountType = accountType;
      UserName = userName;
      Active = active;
      OpenDate = openDate;
      /*Transactions = new List<Transaction>();*/
   }

  /* public void AddTransaction(Transaction transaction)
    {
        Transactions.Add(transaction);
    }*/

   public virtual void GetInfo()
   {
      Console.WriteLine ($"Информация о счете: Номер счета: {AccountNumber}, Баланс: {Balance} рублей, Тип счёта: {AccountType} Имя владельца: {UserName}, Активный счет: {Active} Дата открытия: {OpenDate.ToShortDateString()}");
   }

   public virtual void Deposit(decimal amount)
   {
      Balance += amount;
      Console.WriteLine($"Деньги внесены! Баланс: {Balance} рублей");
   }

   public virtual void Withdraw(decimal amount)
   {
      Balance -= amount;
      Console.WriteLine($"Деньги сняты! Баланс: {Balance} рублей");
   }

    public void CloseScore()
    {
        Active = false;
        Console.WriteLine("Счет закрыт");
    }

}

public class SavingsAccount : BankAccount, IMaxWithdrawalamount
{
   public decimal InterestRate {get; set;}
   public int MaxWithdrawalamount {get; set;}

   public SavingsAccount(int numb, decimal balance, string accountType, decimal interestRate, string userName, bool active, DateTime openDate) : base(numb, balance, accountType,userName, active, openDate)
   {
      
      InterestRate = interestRate;
      MaxWithdrawalamount = 50000;
      
   }
   public override void Deposit(decimal amount)
   {
      Balance += amount * (1 + InterestRate);
      Console.WriteLine($"Деньги внесены с процентной ставкой! Баланс: {Balance} рублей");
   }

    public override void Withdraw(decimal amount)
{
    if (Balance - amount >= MaxWithdrawalamount)
    {
        Balance -= amount;
        Console.WriteLine($"Деньги сняты с учётом минимального баланса");
    }
    else
    {
        Console.WriteLine($"Недостаточно средств на счете для снятия данной суммы");
    }
}

}

public class CheckingAccount : BankAccount, IWithdrawMoney
{
   public decimal OverdraftLimit {get; set;}
   public int WithdrawMoney {get; set;}

   public CheckingAccount(int numb, decimal balance, string accountType, decimal overdraftLimit, string userName, bool active, DateTime openDate) : base(numb, balance, accountType, userName, active, openDate)
   {
      
      OverdraftLimit = overdraftLimit;
      WithdrawMoney = 0;
      
   }
   public override void Withdraw(decimal amount)
   {
      if (Balance + OverdraftLimit >= amount)
        {
            Balance -= amount;
            WithdrawMoney++;
            Console.WriteLine($"Деньги сняты! Баланс: {Balance}, Количество снятий денег:{WithdrawMoney}");
        }
        else
        {
            Console.WriteLine($"Превышен лимит овердрафта");
        }
   }
   
   public class Client<T>
{
    public T Data { get; set; }
    
    public Client(T data)
    {
        Data = data;
    }

    public virtual void ShowData()
    {
        Console.WriteLine(Data);
    }

    protected virtual void ProcessData()
    {
        
    }

    public void RunProcess()
    {
        ProcessData();
        ShowData();
    }
}

}


BankAccount bankAccount1 = new BankAccount(123, 45678, "Дебетовый", "Семен Семеныч", true, DateTime.Now);
bankAccount1.GetInfo();

Console.WriteLine();

SavingsAccount savingsAccount1 = new SavingsAccount(15, 3840, "Дебетовый", 0.03m, "Иван Иванов", false, DateTime.Now);

Console.WriteLine("Сберегательный счёт:");
savingsAccount1.Deposit(300);
savingsAccount1.GetInfo();
savingsAccount1.CloseScore();
Console.WriteLine();

CheckingAccount checkingAccount1 = new CheckingAccount(123, 45678, "Дебетовый", 1000, "Иван Иванов", true, DateTime.Now);
checkingAccount1.Withdraw(70);
Console.WriteLine("Текущий счёт:");
checkingAccount1.GetInfo();
```
